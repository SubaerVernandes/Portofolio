```python
# Ekstraksi Brand
df['brand'] = df['name'].apply(lambda x: x.split()[0])

# Ekstraksi Tipe
df['type'] = df['name'].apply(lambda x: ' '.join(x.split()[1:]).split('(')[0].strip())

# Ekstraksi Warna
df['color'] = df['name'].str.extract(r'\(([^,]+)')[0]

# Ekstraksi RAM (jika tersedia)
df['storage'] = df['name'].str.extract(r', (\d+ GB)\)')[0]

# brand: Mengambil kata pertama dari kolom name, yang diasumsikan sebagai nama merek.
# type: Mengambil teks setelah nama merek hingga sebelum tanda kurung buka, yang berisi informasi tipe atau model.
# color: Mengekstrak warna yang terletak dalam tanda kurung dan sebelum koma pertama.
# RAM: Mengekstrak ukuran RAM yang muncul setelah koma dalam tanda kurung.

df.head()
```

```python
import re
# Fungsi untuk mengekstrak spesifikasi
def extract_specs(corpus):
    if not isinstance(corpus, str):
        return {
            'RAM': 'Not Specified',
            'Sistem Operasi': 'Not Specified',
            'Chipset': 'Not Specified',
            'Processor Speed': 'Not Specified',
            'Kamera': 'Not Specified',
            'Baterai': 'Not Specified',
            'Layar': 'Not Specified',
            'GPU': 'Not Specified',
            'Penyimpanan': 'Not Specified'
        }

    specs = {}

    # Regex patterns untuk berbagai spesifikasi
    patterns = {
        'RAM': r'RAM(\d+ GB)',
        'Sistem Operasi': r'System(Android \d+\.?\d*)',
        'Chipset': r'Processor Type([^\d]+)',
        'Processor Speed': r'Processor Speed(\d+\.\d*)',
        'Kamera': r'(\d+MP)( \d+MP)*',
        'Baterai': r'Capacity(\d+)',
        'Layar': r'Display Size(\d+\.?\d*) cm \(.*\)',
        'GPU': r'GPU([^\d]+)',
        'Penyimpanan': r'Storage(\d+ GB)'
    }

    for key, pattern in patterns.items():
        match = re.search(pattern, corpus)
        if match:
            specs[key] = match.group(1).strip()
        else:
            specs[key] = 'Not Specified'

    return specs

# Terapkan fungsi ke DataFrame
specs_df = df['corpus'].apply(extract_specs).apply(pd.Series)

# Gabungkan dengan DataFrame asli jika diperlukan
df3 = pd.concat([df, specs_df], axis=1)

df3.head()
```

# Build a Model

```python
from sklearn.model_selection import GridSearchCV, train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Definisikan model yang akan di-grid search
model_pipeline = {
    'LinearRegression': {
        'model': Pipeline([
            ('scaler', StandardScaler()),
            ('regressor', LinearRegression())
        ]),
        'params': {
            'regressor__fit_intercept': [True, False],
            'regressor__positive': [True, False]
        }
    },
    'RandomForestRegressor': {
        'model': Pipeline([
            ('scaler', StandardScaler()),
            ('regressor', RandomForestRegressor())
        ]),
        'params': {
            'regressor__n_estimators': [10, 50, 100],
            'regressor__max_depth': [None, 10, 20]
        }
    },
    'SVR': {
        'model': Pipeline([
            ('scaler', StandardScaler()),
            ('regressor', SVR())
        ]),
        'params': {
            'regressor__C': [0.1, 1, 10],
            'regressor__kernel': ['linear', 'rbf']
        }
    }
}

# Fungsi untuk mencari model terbaik menggunakan GridSearchCV
def find_best_model_using_gridsearchcv(x, y):
    scores = []
    for algo_name, config in model_pipeline.items():
        gs = GridSearchCV(config['model'], config['params'], cv=5, return_train_score=False)
        gs.fit(x, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

# Mencari model terbaik
best_model_df = find_best_model_using_gridsearchcv(X_train, y_train)
print(best_model_df)
```

### Berdasarkan ouput diatas kita dapat mengetahui bahwa Random Forest Regressor adalah model dengan skor paling tinggi, maka gunakanlah model tersebut

```python
from sklearn.metrics import mean_squared_error, r2_score

model = RandomForestRegressor(n_estimators=100, random_state=42)

# Melatih model dengan data training
model.fit(X_train, y_train)
```

```python
# Melakukan prediksi pada data uji
y_pred = model.predict(X_test)

# Menghitung Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.4f}')

# Menghitung R-squared (R2)
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2:.4f}')
```
